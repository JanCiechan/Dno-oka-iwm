```python
image=io.imread(plik)
bgr_planes = cv.split(image)
histSize = 256
histRange = (0, 256) # the upper boundary is exclusive
accumulate = False
b_hist = cv.calcHist(bgr_planes, [0], None, [histSize], histRange, accumulate=accumulate)
g_hist = cv.calcHist(bgr_planes, [1], None, [histSize], histRange, accumulate=accumulate)
r_hist = cv.calcHist(bgr_planes, [2], None, [histSize], histRange, accumulate=accumulate)
hist_w = 512
hist_h = 400
bin_w = int(round( hist_w/histSize ))
histImage = np.zeros((hist_h, hist_w, 3), dtype=np.uint8)
cv.normalize(b_hist, b_hist, alpha=0, beta=hist_h, norm_type=cv.NORM_MINMAX)
cv.normalize(g_hist, g_hist, alpha=0, beta=hist_h, norm_type=cv.NORM_MINMAX)
cv.normalize(r_hist, r_hist, alpha=0, beta=hist_h, norm_type=cv.NORM_MINMAX)
for i in range(1, histSize):
    cv.line(histImage, ( bin_w*(i-1), hist_h - int(b_hist[i-1]) ),
            ( bin_w*(i), hist_h - int(b_hist[i]) ),
            ( 255, 0, 0), thickness=2)
    cv.line(histImage, ( bin_w*(i-1), hist_h - int(g_hist[i-1]) ),
            ( bin_w*(i), hist_h - int(g_hist[i]) ),
            ( 0, 255, 0), thickness=2)
    cv.line(histImage, ( bin_w*(i-1), hist_h - int(r_hist[i-1]) ),
            ( bin_w*(i), hist_h - int(r_hist[i]) ),
            ( 0, 0, 255), thickness=2)
#cv.imshow('Source image', image)
#cv.imshow('calcHist Demo', histImage)
#cv.waitKey(0)  
#cv.destroyAllWindows() 
```

```python
def calosc(plik='HRF/healthy/01_h.jpg',plik2='HRF/healthy_manualsegm/01_h.tif',plik3='HRF/healthy_fovmask/01_h_mask.tif',remove_small=False): 
    image=io.imread(plik)
    R, G, B = cv.split(image)

    output1_R = R#cv.equalizeHist(R)
    output1_G = cv.equalizeHist(G)
    output1_B = B#cv.equalizeHist(B)

    image = cv.merge((output1_G, output1_G, output1_G))
    image=cv.cvtColor(image, cv.COLOR_RGB2GRAY)
    blur = cv.GaussianBlur(image,(15,15),23)
    weight=cv.addWeighted(image, 1.5, blur, -0.5, 0, blur);
    frangi_result=frangi(weight)
    img3=cv.imread(plik3,0)
    img1=cv.imread(plik,cv.IMREAD_GRAYSCALE)
    resulting_img=frangi_result*img3
    sum=0
    max=0
    min=1
    for i in range(len(resulting_img)):
        for j in range(len(resulting_img[0])):
            if(resulting_img[i][j]>max):
                max=resulting_img[i][j]
            if(resulting_img[i][j]<min):
                min=resulting_img[i][j]
    for i in range(len(resulting_img)):
        for j in range(len(resulting_img[0])):
            resulting_img[i][j]=(resulting_img[i][j]-min)/(max-min)
            resulting_img[i][j]= resulting_img[i][j]*255

    kernel = np.ones((5,5),np.float32)/25
    resulting_img = cv.filter2D(resulting_img,-1,kernel)
    ret,thresh = cv.threshold(resulting_img,200,255,cv.THRESH_BINARY)
    kernel = np.ones((25, 25), np.uint8)
    erode = cv.erode(thresh, kernel, iterations = 10)
    resulting_img = cv.bitwise_or(resulting_img, erode)
    for i in range(len(resulting_img)):
        for j in range(len(resulting_img[0])):
            if(resulting_img[i][j]>5):
                resulting_img[i][j]=255
            else:
                resulting_img[i][j]=0
    removed_objects=resulting_img/255
    if(remove_small):
        for i in range(len(removed_objects)):
            for j in range(i):
                if(removed_objects[i][j]!=1):
                    removed_objects[i][j]=0
        removed_objects=removed_objects.astype(bool)

        removed_objects=morphology.remove_small_objects(removed_objects, min_size=600, connectivity=5)
        
    handdrawn=cv.imread(plik2,0)
    handdrawn=handdrawn.astype(bool)
    
    macierz_pomylek=np.zeros((len(handdrawn),len(handdrawn[0])),dtype=bool)
    zgodnosc=0
    niezgodnosc=0
    maska=io.imread(plik3, as_gray=True)
    for i in range(len(handdrawn)):
        for j in range(len(handdrawn[0])):
            if(maska[i][j]==1):
                if(handdrawn[i][j]!=removed_objects[i][j]):
                    macierz_pomylek[i][j]=1
                else:
                    macierz_pomylek[i][j]=0
                    if(handdrawn[i][j]==1):
                        zgodnosc+=1
                    else:
                        niezgodnosc+=1
    Sensitivity=zgodnosc/handdrawn.sum()
    #print("Sensitivity wynosi:",Sensitivity*100,"%")#true positive
    Specificity=niezgodnosc/(handdrawn.size-(maska.size-maska.sum())-handdrawn.sum())
    #print("Specificity wynosi:",Specificity*100,"%")
    Accuracy=(zgodnosc+niezgodnosc)/(handdrawn.size-(maska.size-maska.sum()))
    #print("Accuracy wynosi:",Accuracy*100,"%")#(tp+tn)/P+N
    image=io.imread(plik)
    return image,handdrawn,removed_objects,macierz_pomylek,Sensitivity,Specificity,Accuracy
```

```python
imgs=[]
handdrawns=[]
masks=[]
for i in range(1,6):
    img='HRF/healthy/0'+str(i)+'_h.jpg'
    imgs.append(img)
    expert='HRF/healthy_manualsegm/0'+str(i)+'_h.tif'
    handdrawns.append(expert)
    mask='HRF/healthy_fovmask/0'+str(i)+'_h_mask.tif'
    masks.append(mask)
w=10
h=10
fig=plt.figure(figsize=(20, 20))
columns = 4
rows = 5
x=0
licznik=1
for i in range(1, 6):
    image,handdrawn,removed_objects,macierz_pomylek,Sensitivity,Specificity,Accuracy=calosc(plik=imgs[x],plik2=handdrawns[x],plik3=masks[x])
    fig.add_subplot(rows, columns, licznik)
    plt.title('Original')
    plt.imshow(image)
    licznik+=1
    fig.add_subplot(rows, columns, licznik)
    plt.title('Handdrawn')
    plt.imshow(handdrawn)
    licznik+=1
    fig.add_subplot(rows, columns, licznik)
    plt.title('After changes')
    plt.imshow(removed_objects)
    licznik+=1
    fig.add_subplot(rows, columns, licznik)
    textstr='Sensitivity: '+str(round(Sensitivity*100, 2))+'%\n'+'Specificity: '+str(round(Specificity*100, 2))+'%\n'+'Accuracy: '+str(round(Accuracy*100, 2))+'%\n'
    plt.text(2, .4, textstr,
        bbox=dict(boxstyle = "square",
                  facecolor = "white"))
    plt.title('Mistake matrix')
    plt.imshow(macierz_pomylek,cmap='gray')
    licznik+=1
    x+=1
plt.show()
```
